##1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.

- Descriptive statistics summarize and describe the features of a dataset, such as mean, median, and charts. They help us understand what the data looks like.
Inferential statistics use samples to make predictions or decisions about a larger population.
For example, calculating the average marks of a class is descriptive, but predicting the performance of the entire school based on one class is inferential.

##2. What is sampling in statistics? Explain the differences between random and stratified sampling.

- Sampling means selecting a small part of a population to analyze instead of studying the whole group.
In random sampling, every individual has an equal chance of selection.
Stratified sampling divides the population into groups (strata) like age or gender, and samples are taken from each group.
Stratified sampling gives better representation when the population has distinct categories.

##3. Define mean, median, and mode. Explain why these measures of central tendency are important.

- Mean is the average of all values, median is the middle value when data is sorted, and mode is the most repeated value.
These measures help identify the center or typical value in a dataset.
They also help compare different datasets and understand data distribution.
Each measure works better for different types of data, especially when outliers exist.

##4. Explain skewness and kurtosis. What does a positive skew imply about the data?

- Skewness measures how stretched or asymmetric a distribution is.
Kurtosis measures the "tailedness" or how extreme the outliers are.
A positive skew means the tail is longer on the right side, showing more high extreme values.
It also means most data values are on the left, with few large values pulling the mean upward.

##5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.
- Numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


✅ Python Code

```python
import numpy as np
from scipy import stats

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = np.mean(numbers)
median_value = np.median(numbers)
mode_value = stats.mode(numbers, keepdims=True).mode[0]

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

##6. Compute the covariance and correlation coefficient between the following two datasets.
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

✅ Python Code

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y)[0][1]
correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation:", correlation)
```

##7. Write a Python script to draw a boxplot for the numeric list and identify its outliers. Explain the result.
  - data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

  Explanation

  - The boxplot shows the median, quartiles, and possible outliers.
  - In this dataset, 35 appears as a higher value that stretches the whisker and may be treated as an outlier.
  - Most values lie between 14 and 26, showing the data is moderately spread out.
The distribution is slightly right-skewed.

  - ✅ Python Code

```python
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.show()
```

##8. Relationship between advertising spend and daily sales using covariance and correlation

- Covariance shows whether two variables move together, while correlation shows how strongly and in what direction they move together.
A positive covariance means both increase together. A high correlation close to 1 means strong positive relationship.
This helps the marketing team understand if higher ad spending leads to higher sales.

  - advertising_spend = [200, 250, 300, 400, 500]

  - daily_sales = [2200, 2450, 2750, 3200, 4000]
  
✅ Python Code

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

correlation = np.corrcoef(advertising_spend, daily_sales)[0][1]

print("Correlation:", correlation)
```

##9. Customer satisfaction scores (1–10 scale)

- To understand distribution, we use measures like mean, standard deviation, and median.

    survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
